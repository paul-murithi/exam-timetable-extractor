```python
import pandas as pd

my_units = ["BAF1101", "BBM1101", "BBM1202", "BIT2103", "BMA1104", "BMA1106", "BPY1101", "BUCU007"]
file_path = "Final-Bachelor's-Virtual-Exam-Timetable-Dec 2025.xlsx"
```

```python
print("Reading Excel file and filtering units...")
df = pd.read_excel(file_path, header=5)
```

```python
print("Filtering for specified course codes...")
filtered = df[df["Course_Code"].isin(my_units)]

clean = filtered[["Exams Date", "Session Time", "Course_Code", "Course_Title"]].copy()
is_date = clean["Exams Date"].astype(str).str.match(r'^\d{4}-\d{2}-\d{2}$')
clean.loc[is_date, "Exams Date"] = pd.to_datetime(clean.loc[is_date, "Exams Date"], errors='coerce').dt.date
print("Done!")
print(clean)
```
